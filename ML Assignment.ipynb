## Splitting Data into Train, Test & Split

```python
X = df.drop('TARGET_deathRate',axis=1)  # Independent 
y = df['TARGET_deathRate']  # Dependent variable
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y , train_size=0.7,test_size = 0.3, random_state = 0)
```

```python
X_train.shape
```

```python
X_test.shape
```

## Feature Selection

```python
from sklearn.linear_model import Lasso  # Doing feature selection with the help of lasso regression.
from sklearn.feature_selection import SelectFromModel
```

```python
feature_selection = SelectFromModel(Lasso(alpha = 0.05, random_state=0))
feature_selection.fit(X_train, y_train)
```

```python
feature_selection.get_support()
```

```python
selected_feat = X_train.columns[(feature_selection.get_support())]
```

```python
selected_feat # All features are selected at alpha = 0.05
```

```python
X_train = X_train[selected_feat]
X_test = X_test[selected_feat]
```

## Hyperparameter Tuning & Applying Model

```python
from sklearn.model_selection import RandomizedSearchCV  # Applying RandomizedSearchCV for faster computation.

n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1200, num = 10)]
max_features = ["auto", "sqrt", "log2"]
max_depth = [int(x) for x in np.linspace(6, 24, num = 6)]
min_samples_split = [2, 5, 10, 15, 100]
min_samples_leaf = [1, 2, 5, 10]


random_grid = {'n_estimators' : n_estimators,
              'max_features' : max_features,
              'max_depth' : max_depth,
              'min_samples_split' : min_samples_split,
              'min_samples_leaf' : min_samples_leaf}
print(random_grid)
```

```python
from sklearn.ensemble import RandomForestRegressor  # Applying Random Forest Regression as ML Technique.
rf = RandomForestRegressor()

rf_random = RandomizedSearchCV(estimator = rf, param_distributions=random_grid, scoring='neg_mean_squared_error', n_iter = 25, cv = 4)
```

```python
rf_random.fit(X_train, y_train)  # Fitting the Training Data
y_pred = rf_random.predict(X_test)  # predicting with test data
```

## Checking the accuracy of the model. 

```python
from sklearn.metrics import mean_absolute_error as MAE # Mean Abs Error.
from sklearn.metrics import mean_squared_error as MSE # Mean square Error.

rmse = (MSE(y_pred, y_test))**(1/2)  # Root mean squared error.

print('MAE ->', MAE(y_pred, y_test))
print('MSE ->', MSE(y_pred, y_test))
print('RMSE ->', rmse)
```
